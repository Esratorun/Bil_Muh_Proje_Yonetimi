## **Ä°Ã§erik ve GÃ¶revler**

- **AmaÃ§ Fonksiyonu (amac_fonksiyonu)**  
  Pigment oranlarÄ±na gÃ¶re renk kalitesi puanÄ±nÄ± hesaplar.

- **KÄ±sÄ±t KontrolÃ¼ (kisit_kontrol)**  
  Bireylerin problemde tanÄ±mlanan kÄ±sÄ±tlarÄ± saÄŸlayÄ±p saÄŸlamadÄ±ÄŸÄ±nÄ± kontrol eder.

- **Uygunluk HesabÄ± (uygunluk_hesapla)**  
  AmaÃ§ fonksiyonu ve kÄ±sÄ±t cezalarÄ±nÄ± kullanarak fitness deÄŸerini hesaplar.

- **SeÃ§im OperatÃ¶rleri**  
  - rulet_secimi : Uygunluk deÄŸerine baÄŸlÄ± olasÄ±lÄ±ksal seÃ§im yapar.  
  - rank_temelli_secim : Bireyleri sÄ±ralamaya gÃ¶re seÃ§er.

- **Ã‡aprazlama (tek_noktali_caprazlama)**  
  SeÃ§ilen ebeveynlerden yeni bireyler Ã¼retir.

- **Mutasyon (mutasyon_uygula)**  
  PopÃ¼lasyonda genetik Ã§eÅŸitliliÄŸi artÄ±rmak iÃ§in rastgele deÄŸiÅŸimler uygular.

- **PopÃ¼lasyon OluÅŸturma (populasyon_olustur)**  
  BaÅŸlangÄ±Ã§ popÃ¼lasyonunu kÄ±sÄ±tlara uygun ÅŸekilde rastgele oluÅŸturur.

---

Bu dosya, genetik algoritmanÄ±n **evrimsel sÃ¼recinin temelini** oluÅŸturur.


```python
kod='''
import numpy as np
import random

karsilik_cezasi =-5000 # KÄ±sÄ±t ihlallerinde verilecek ceza (Uygunluk deÄŸerinden dÃ¼ÅŸecek)
gen_sayisi=2

# KULLANILACAK TEMEL FONKSÄ°YONLAR (AmaÃ§ ve Uygunluk FonksiyonlarÄ±)

"""
AMAÃ‡ FONKSÄ°YONU (Renk Kalitesi PuanÄ± ) :   y = 5*x1 + 2*x2 - x1*x2

        birey (np.ndarray): [x1, x2] deÄŸerlerini iÃ§eren kromozom parametre olarak alÄ±nÄ±r.

        float: Hesaplanan y (Renk Kalitesi) puanÄ± dÃ¶ndÃ¼rÃ¼lÃ¼r.

"""

def amac_fonksiyonu(birey: np.ndarray) -> float:

    x1, x2 = birey
    # AmaÃ§ fonksiyonu: y = 5 * x1 + 2 * x2 - x1 * x2 (Maksimize edilecek)
    y = 5 * x1 + 2 * x2 - x1 * x2
    return y


"""
    Bireyin kÄ±sÄ±tlarÄ± ihlal edip etmediÄŸini kontrol eder.

    KÄ±sÄ±tlar:
    1. x1 + x2 = 100
    2. x1 >= 30
    3. x1 : [0,100] ve  x2 : [0,100]

        birey (np.ndarray): [x1, x2] deÄŸerlerini iÃ§eren birey kromozomu parametre alÄ±r.
        int: Ä°hlal edilen kÄ±sÄ±t sayÄ±sÄ± (0, 1 veya 2) dÃ¶ndÃ¼rÃ¼r.

 """

def kisit_kontrol(birey: np.ndarray) -> int:
    x1, x2 = birey
    ihlal_sayisi=0

    # x1 + x2 = 100 olmalÄ± ama float kullandÄ±ÄŸÄ±mÄ±z iÃ§in kÃ¼Ã§Ã¼k bir tolerans(epilon) kullanÄ±yoruz.

    epsilon=1e-4

    if abs( x1+x2-100)>epsilon:
        ihlal_sayisi+=1

    # x1 >= 30 olmalÄ±

    if x1<30:
        ihlal_sayisi+=1

    # x1 : [0,100] ve  x2 : [0,100]

    if not (0 <= x1 <= 100 and 0 <= x2 <= 100):
        ihlal_sayisi += 1


    return ihlal_sayisi


"""
    Bireyin uygunluk (fitness) skorunu hesaplar.
    Bu, amaÃ§ fonksiyonu sonucundan kÄ±sÄ±t cezalarÄ±nÄ±n dÃ¼ÅŸÃ¼lmÃ¼ÅŸ halidir.

        birey (np.ndarray): [x1, x2] deÄŸerlerini iÃ§eren birey kromozomu parametre alÄ±r.
        float: Nihai uygunluk skoru dÃ¶ndÃ¼rÃ¼r.

"""

def uygunluk_hesapla(birey: np.ndarray) -> float:

    y_puan =amac_fonksiyonu(birey)
    ihlal_sayisi=kisit_kontrol(birey)

    #kisit_kontrol fonksiyonundan gelen ihlal sayÄ±sÄ± ile baÅŸta tanÄ±mladÄ±ÄŸÄ±m karsilik_cezasi kullanarak ceza hesaplanÄ±r.

    ceza = ihlal_sayisi*karsilik_cezasi

    uygunluk = y_puan + ceza

    return uygunluk


# SEÃ‡Ä°M OPERATÃ–RLERÄ° (Rulet ve Rank)

    """
    RULET SEÃ‡Ä°M METODU  :

    Uygunluk deÄŸerlerine gÃ¶re rulet tekeri yÃ¶ntemiyle ebeveyn seÃ§er.
    UygunluÄŸu yÃ¼ksek olan bireylerin seÃ§ilme olasÄ±lÄ±ÄŸÄ± daha fazladÄ±r.

    """

def rulet_secimi(populasyon: list, uygunluklar: np.ndarray, adet: int = 2) -> np.ndarray:
    # OlasÄ±lÄ±k hesaplamasÄ± iÃ§in tÃ¼m uygunluklar pozitif olmalÄ±.
    # Bunun iÃ§in en kÃ¼Ã§Ã¼k uygunluk deÄŸerine kaydÄ±rma uyguluyoruz.
    min_uygunluk = np.min(uygunluklar)

    # TÃ¼m uygunluklar pozitif olacak ÅŸekilde offset eklenir.
    offset = np.abs(min_uygunluk) + 1e-6
    pozitif_uygunluklar = uygunluklar + offset

    toplam_uygunluk = np.sum(pozitif_uygunluklar)
    olasiliklar = pozitif_uygunluklar / toplam_uygunluk

    # Birey listesini numpy array'e Ã§eviriyoruz.
    populasyon_np = np.array(populasyon, dtype=object)

    secilen_indeksler = np.random.choice(
        len(populasyon_np),
        size=adet,
        p=olasiliklar,
        replace=False
    )

    return populasyon_np[secilen_indeksler]


    """

    RANK SEÃ‡Ä°M METODU  :

    Bireyler uygunluk deÄŸerlerine gÃ¶re sÄ±ralanÄ±r, sÄ±ralamadaki yerlerine gÃ¶re seÃ§im olasÄ±lÄ±ÄŸÄ± atanÄ±r.

    """

def rank_temelli_secim(populasyon: list, uygunluklar: np.ndarray, adet: int = 2) -> np.ndarray :

     N = len(populasyon)

     sirali_indeksler = np.argsort(-uygunluklar) # BÃ¼yÃ¼kten kÃ¼Ã§Ã¼ÄŸe sÄ±ralÄ±yoruz

     # PopÃ¼lasyonu numpy dizisine Ã§eviriyoruz
     populasyon_np = np.array(populasyon, dtype=object)

     payda = N * (N + 1) / 2
     secim_ihtimalleri = np.array([(N - i) / payda for i in range(N)]) #en iyi bireye en yÃ¼ksek lasÄ±lÄ±k deÄŸeri verilmeli

     secilen_indeksler = np.random.choice(
        N,
        size=adet,
        p=secim_ihtimalleri,
        replace=False
       )

      # SeÃ§ilen sÄ±ralÄ± indeksler kullanÄ±larak gerÃ§ek bireyler alÄ±nÄ±r.Yani sirali_indeksler hangi bireye karÅŸÄ±lÄ±k geliyorsa o alÄ±nÄ±yor.
     ebeveynler = [populasyon_np[sirali_indeksler[i]] for i in secilen_indeksler]

     return np.array(ebeveynler)



#Ã‡APRAZLAMA OPERATÃ–RLERÄ°

"""
TEK VE Ä°KÄ° NOKTALI Ã‡APRAZLAMA :

    Rastgele bir noktadan ebeveynlerin kromozomlarÄ±nÄ± kesip Ã§aprazlar.
    Gen sayÄ±sÄ± 2 (x1, x2) olduÄŸu iÃ§in kesim noktasÄ± sabittir (x1 ile x2 arasÄ±).
    Ä°ki noktalÄ± Ã§aprazlamada ise kesme noktasÄ± rastgele belirlenir.Fakat bizim gen sayÄ±mÄ±z iki olduÄŸu iÃ§in buna gerek yok.
    Projede sadece tek noktalÄ± Ã§aprazlama kullanacaÄŸÄ±m.

"""

def tek_noktali_caprazlama(p1: np.ndarray, p2: np.ndarray) -> tuple[np.ndarray, np.ndarray]:

    # Kesim noktasÄ± 1 yani x1'den sonra
    nokta = 1

    c1 = np.concatenate((p1[:nokta], p2[nokta:]))
    c2 = np.concatenate((p2[:nokta], p1[nokta:]))

    return c1, c2


# MUTASYON OPERATÃ–RÃœ

"""
    Her gen iÃ§in rastgele mutasyon uygulayarak Ã§Ã¶zÃ¼m Ã§eÅŸitliliÄŸini artÄ±rÄ±r.
    Mutasyon sonrasÄ± deÄŸerler [0, 100] aralÄ±ÄŸÄ±na kÄ±sÄ±tlanÄ±r.
"""

def mutasyon_uygula(birey: np.ndarray, ihtimal: float, buyukluk: float) -> np.ndarray:

  yeni = birey.copy()

      # Her gen iÃ§in mutasyon olasÄ±lÄ±ÄŸÄ± kontrol edilir
  for i in range(len(yeni)):
        if np.random.rand() < ihtimal:
            # Rastgele bir deÄŸiÅŸim miktarÄ± hesaplama
            degisim = buyukluk * (np.random.rand() - 0.5) * 2  #Rastgele sapma 1 ve -1 arasÄ±nda olsun
            yeni[i] += degisim

            # Gen deÄŸerleri belirlenen sÄ±nÄ±rlar iÃ§inde tutmak iÃ§in clip
            yeni[i] = np.clip(yeni[i], 0, 100)


  return yeni



# POPÃœLASYON OLUÅžTURMA

"""
  BaÅŸlangÄ±Ã§ popÃ¼lasyonunu rastgele oluÅŸturur.
  x1 ve x2 deÄŸerleri 0-100 aralÄ±ÄŸÄ±nda olmalÄ±dÄ±r.

  x1 + x2 = 100 kÄ±sÄ±tÄ±nÄ± saÄŸlamak iÃ§in normalizasyon yapÄ±lÄ±r.
"""



def populasyon_olustur(boyut: int, gen_sayisi: int) -> list:
    populasyon = []

    for i in range(boyut):
        # Rastgele bireyler oluÅŸturuluyor
        bire = np.random.uniform(0, 100, gen_sayisi)

        # Normalizasyon
        toplam = np.sum(bire)

        # x1, x2 ... deÄŸerlerinin toplamÄ± 100 olacak ÅŸekilde Ã¶lÃ§ekleme
        if toplam > 0:
            oran = 100 / toplam
            bire *= oran

        # DeÄŸerler [0, 100] aralÄ±ÄŸÄ±nda kalsÄ±n
        bire = np.clip(bire, 0, 100)

        populasyon.append(bire)

    return populasyon

'''

# DOSYAYA YAZMA

with open ('genetik_operators.py', 'w') as dosya:
    dosya.write(kod)
print("genetik_operators.py dosyasÄ± baÅŸarÄ±yla oluÅŸturuldu.")
```

# **evrim_motoru.py DOSYASI**

Bu dosya, Genetik AlgoritmanÄ±n **ana Ã§alÄ±ÅŸma motorunu** iÃ§erir ve tÃ¼m evrimsel sÃ¼reci yÃ¶netir.

---

## Ä°Ã§erik ve GÃ¶revler

- **KÄ±sÄ±t DÃ¼zeltme (kisit_duzelt)**  
  Bireylerin x1 + x2 = 100 kÄ±sÄ±tÄ±nÄ± saÄŸlamasÄ± iÃ§in normalizasyon uygular.

- **Genetik Optimizasyon Motoru (genetik_optimizasyon_motoru)**  
  Genetik algoritmanÄ±n tÃ¼m adÄ±mlarÄ±nÄ± sÄ±rasÄ±yla gerÃ§ekleÅŸtirir:
  
  - BaÅŸlangÄ±Ã§ popÃ¼lasyonunun oluÅŸturulmasÄ±  
  - Uygunluk (fitness) hesaplamasÄ±  
  - Elitizm ile en iyi bireyin korunmasÄ±  
  - Rulet veya rank tabanlÄ± seÃ§im  
  - Ã‡aprazlama ve mutasyon iÅŸlemleri  
  - KÄ±sÄ±t ihlallerinin dÃ¼zeltilmesi  

- **Evrim SÃ¼reci Takibi**  
  Her nesilde en iyi birey ve uygunluk deÄŸeri ekrana yazdÄ±rÄ±lÄ±r.

- **GÃ¶rselleÅŸtirme**  
  Nesiller boyunca en iyi uygunluk deÄŸerlerinin deÄŸiÅŸimi grafik olarak Ã§izilir ve dosyaya kaydedilir.

---

Bu dosya, genetik algoritmanÄ±n **kontrol merkezi** olup optimizasyon sÃ¼recinin tamamÄ±nÄ± koordine eder.


```python
kod_motor = '''
import numpy as np
import random
import matplotlib.pyplot as plt
from genetik_operators import (
    amac_fonksiyonu,
    kisit_kontrol,
    uygunluk_hesapla,
    rulet_secimi,
    rank_temelli_secim,
    tek_noktali_caprazlama,
    mutasyon_uygula,
    populasyon_olustur,
    gen_sayisi
)



# YARDIMCI DÃœZELTME FONKSÄ°YONU
def kisit_duzelt(birey: np.ndarray) -> np.ndarray:
    """x1 + x2 = 100 kÄ±sÄ±tÄ±nÄ± saÄŸlamak iÃ§in bireyi normalleÅŸtirir."""
    toplam = np.sum(birey)

    # abs(toplam - 100) > 1e-4 kontrolÃ¼, kÄ±sÄ±tÄ±n ihlal edilip edilmediÄŸini kontrol eder
    if toplam > 0 and abs(toplam - 100) > 1e-4:
        oran = 100 / toplam
        birey *= oran

    # DeÄŸerlerin 0-100 aralÄ±ÄŸÄ±nda kalmasÄ±nÄ± saÄŸla
    return np.clip(birey, 0, 100)






def genetik_optimizasyon_motoru(
    populasyon_boyutu: int,
    nesil_sayisi: int,
    caprazlama_turu: str,
    secim_turu: str,
    mutasyon_ihtimali: float,
    mutasyon_buyuklugu: float
):

    # ADIM 1 :BAÅžLANGIÃ‡ POPÃœLASYONU OLUÅžTURMA

    populasyon = populasyon_olustur(populasyon_boyutu,gen_sayisi)
    en_iyi_uygunluk_gecmisi = []
    en_iyi_birey = None
    en_yuksek_uygunluk = -np.inf

    print(f"GENETÄ°K ALGORÄ°TMA BAÅžLATILDI ({secim_turu.upper()} SeÃ§im)")

    for nesil in range(nesil_sayisi):

        # ADIM 2 UYGUNLUK (FÄ°TNESS) HESAPLAMA
        uygunluklar = np.array([uygunluk_hesapla(birey) for birey in populasyon])

        # ADIM 3 EÄ°LTÄ°ZM Ä°LE EN Ä°YÄ° BÄ°REYÄ° BULMA
        en_iyi_indeks = np.argmax(uygunluklar)
        mevcut_elit_birey = populasyon[en_iyi_indeks].copy()
        mevcut_elit_uygunluk = uygunluklar[en_iyi_indeks]

        # Genel en iyi bireyi (Global Best) gÃ¼ncelleme
        if mevcut_elit_uygunluk > en_yuksek_uygunluk:
            en_yuksek_uygunluk = mevcut_elit_uygunluk
            en_iyi_birey = mevcut_elit_birey.copy()

        # GeÃ§miÅŸe kaydetme (Global Best'i kaydetmek en doÄŸru yaklaÅŸÄ±mdÄ±r)
        en_iyi_uygunluk_gecmisi.append(en_yuksek_uygunluk)

        # EKRANDA GÃ–STERELÄ°M (Ã‡Ä±ktÄ± formatÄ± dÃ¼zeltildi)
        print(f"\\n 1. Nesil: {nesil+1}/{nesil_sayisi}")
        print(f"2. En Ä°yi Birey (x1, x2): [{en_iyi_birey[0]:.3f}, {en_iyi_birey[1]:.3f}]")
        print(f"3. Renk Kalitesi (y): {amac_fonksiyonu(en_iyi_birey):.4f}")
        print(f"4. Uygunluk (Fitness): {en_yuksek_uygunluk:.4f}")
        print(f"5. KÄ±sÄ±t Ä°hlali: {kisit_kontrol(en_iyi_birey)}")

        # ADIM 4 EÅžLEÅžME HAVUZU OLUÅžTURMA
        havuz=[]

        for _ in range(populasyon_boyutu // 2):
            if secim_turu == "rulet":
                ebeveynler = rulet_secimi(populasyon, uygunluklar)
            elif secim_turu == "rank":
                ebeveynler = rank_temelli_secim(populasyon, uygunluklar)
            else:
                raise ValueError("SeÃ§im tÃ¼rÃ¼ 'rulet' veya 'rank' olmalÄ±.")
            havuz.append(ebeveynler)

        # ADIM 5 Ã‡APRAZLAMA VE MUTASYON
        yeni_populasyon=[en_iyi_birey]

        while len(yeni_populasyon)<populasyon_boyutu:
            p1,p2=random.choice(havuz)

            if caprazlama_turu=="tek" or caprazlama_turu == "iki":
                c1,c2=tek_noktali_caprazlama(p1,p2)
            else:
                raise ValueError("Ã‡aprazlama tÃ¼rÃ¼ 'tek' veya 'iki' olmalÄ±.")

            c1=mutasyon_uygula(c1,mutasyon_ihtimali,mutasyon_buyuklugu)
            c2=mutasyon_uygula(c2,mutasyon_ihtimali,mutasyon_buyuklugu)

            # ðŸš¨ EKLENMESÄ° GEREKEN KOD BLOÄžU BAÅžLANGICI: DÃœZELTME
            c1 = kisit_duzelt(c1)
            c2 = kisit_duzelt(c2)

            yeni_populasyon.extend([c1,c2])

        populasyon=np.array(yeni_populasyon[:populasyon_boyutu])

    print("\\n---  OPTÄ°MÄ°ZASYON TAMAMLANDI ---\\n")

    # Grafik Ã§izimi
    plt.figure(figsize=(10, 6))
    plt.plot(en_iyi_uygunluk_gecmisi, marker='.', linestyle='-', color='indigo')
    # SatÄ±r 60 civarÄ±ndaki f-string hatasÄ±nÄ±n kaynaÄŸÄ± olabilecek tÄ±rnaklar kontrol edildi.
    plt.title(f"Genetik Algoritma Evrimi: En Ä°yi Uygunluk ({secim_turu.upper()} SeÃ§im)")
    plt.xlabel("Nesil SayÄ±sÄ±")
    plt.ylabel("En Ä°yi Uygunluk DeÄŸeri")
    plt.grid(True)
    plt.tight_layout()

    # ðŸš¨ EKLENECEK KOMUTLAR: GrafiÄŸi dosyaya kaydetme YAPAY ZEKA Ä°LE DÃœZELTÄ°LDÄ°
    try:
        # GrafiÄŸi 'evrim_grafiÄŸi.png' adÄ±yla kaydediyoruz
        plt.savefig('evrim_grafiÄŸi.png')
    except Exception as e:
        # Hata olursa ekrana yazdÄ±r
        print(f"HATA: Grafik kaydedilemedi: {e}")

    # BelleÄŸi temizler, bÃ¶ylece notebook ortamÄ± yorulmaz
    plt.close()

    return en_iyi_birey,en_yuksek_uygunluk,en_iyi_uygunluk_gecmisi
'''

with open("evrim_motoru.py", "w") as dosya:
    dosya.write(kod_motor)

```
